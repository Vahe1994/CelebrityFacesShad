```python
import face_recognition
import matplotlib.pyplot as plt
import numpy as np
from imageio import imread

def get_face(img=None,url=None):
    face_locations=None
    if img is None:
        img = face_recognition.load_image_file(url)
        face_locations = face_recognition.face_locations(img)
    else:
        face_locations = face_recognition.face_locations(img)
    img =img[list(face_locations[0])[0]:list(face_locations[0])[2],list(face_locations[0])[3]:list(face_locations[0])[1]]
    return img
```

```python
path = 'img_align_celeba'
```

```python
a = listdir(path)
```

```python
a.sort()
file = open('mass_with_emb_cropped.txt','w')
j=0
for it in tqdm(a):
    try:
        img = get_face(img=None,url= path+'/'+it) 
        scipy.misc.imsave('lalala.jpg', img)
        image = face_recognition.load_image_file('lalala.jpg')
        enc = face_recognition.face_encodings(image)[0]
        st= str(it)+'\t'
        i=0
        for it_1 in enc:
            if(i == enc.shape[0]-1):
                st += str(it_1)+'\n'
            else:
                st += str(it_1)+' '
            i+=1
        file.write(st)
       
    except:
        continue
    j+=1
file.close()
```
